```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm
import scipy.stats as stat
%matplotlib inline 
```

```python
# generate grid points for joint pdf evaluation
gridpoints = 200
xpoints, ypoints = np.meshgrid(np.linspace(-5,5,num=gridpoints),np.linspace(-5,5,gridpoints))

# put the x-y coordinates into a grid
grid = np.empty(xpoints.shape + (2,))
grid[:, :, 0] = xpoints
grid[:, :, 1] = ypoints

# helper function for calculating correlation coefficient
def get_cc(cov_mat):
    # 0,1 -> cov[x,y]; 0,0 -> var_x; 1,1 -> var_y
    corr_coeff = float(cov_mat[0][1])/math.sqrt(cov_mat[0][0]*cov_mat[1][1])
    print('The correlation coefficient is ' + str(corr_coeff))
```

```python
# part a) Plot contours for mean value [0,0], covariance identity.  
# Note the plot command to set the aspects so that the x-axis and the y-axis have the same length!

# evaluate joint Gaussian pdf at grid points
mean_vector = [0,0]
covariance_matrix = [[1,0],[0,1]]
fXY = stat.multivariate_normal.pdf(grid,mean_vector,covariance_matrix); 

fig = plt.figure()
plt.contourf(xpoints, ypoints, fXY, cmap = 'cool', levels=7)
plt.colorbar()
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.draw()

get_cc(covariance_matrix)
fig.savefig(f'part_a.png')
```
